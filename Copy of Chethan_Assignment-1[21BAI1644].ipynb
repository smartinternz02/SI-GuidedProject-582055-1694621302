## **Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)**

```python
import pandas as pd
import numpy as np
import random
random.seed(42)
ages = [np.random.randint(18, 45) for _ in range(4)]
Salaries = [np.random.randint(50000, 800000) for _ in range(4)]
cities= ['Mumbai', 'Delhi', 'Kolkata', 'Hyderabad', 'Bangalore', 'Chennai']
City = [np.random.choice(cities) for _ in range(4)]
Degree=['Bachelor', 'Master', 'BBA', 'Doctor']
Education=[np.random.choice(Degree) for _ in range(4)]
data = {
    'Age': ages,
    'Salary': Salaries,
    'Education Level': Education,
    'Living City': City,

}
df = pd.DataFrame(data)
df

```

# **Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4**

```python
df.rename(columns={'Age':'Random value 1','Salary':'Random value 2','Education Level':'Random value 3','Living City':'Random value 4'},inplace=True)
df
```
